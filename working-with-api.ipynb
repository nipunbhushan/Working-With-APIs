### Dog API by Kinduff

```python
url="http://dog-api.kinduff.com/api/facts?number="+str(randint(1,100))
response = requests.get(url).json()
for fact in response['facts']:
    print("Fact:")
    print(fact)
```

### Open Library API

List of fantasy book from year 2010-2022

```python
url="http://openlibrary.org/subjects/fantasy.json?details=true&published_in==2010-2022"
response = requests.get(url)
response.status_code
```

```python
pd.DataFrame.from_dict(requests.get(url).json()['works'])
```

List of science book from year 1800-1900 and have ebooks available

```python
url="http://openlibrary.org/subjects/science.json?details=true&published_in==1800-1900&ebooks=true"
response = requests.get(url)
response.status_code
```

```python
pd.DataFrame.from_dict(requests.get(url).json()['works'])
```
